 ### 1. Columna de species:
  si exploramos esta columna en detalle nos podemos dar cuenta que tenemos muchos valores 煤nicos y esto hace que sea muy dif铆cil trabajar con esta columna. Lo que vamos a hacer es clasificar los tiburones en 5 especies diferentes, las m谩s comunes, que incluyen el tibur贸n blanco (White), el tibur贸n tigre (Tiger ), el tibur贸n gris (Grey), el tibur贸n lim贸n (Lemon) y el tibur贸n toro (Bull). El resto de las especies las incluiremos en un 煤nico grupo que podremos llamar "Unspecified". Ahora nos podemos sentir un poco abrumadas y no saber como enfrentarnos a este reto, pero don't worry, os dejamos por aqu铆 unas  pistas  para que os ayuden a entender c贸mo hacerlo.

* Los valores de las columnas son strings por lo que podremos usar regex para buscar palabras clave en cada celda y asignarlo a una de las categor铆as que hemos definido previamente.

```python
df = pd.read_csv("datos/1.ataques_limpio.csv", index_col=0)
```

```python
df.head(2)
```

```python
df["species_"].unique()
```

```python
def cambiar_especie(col):

    patron_blanco= r".*[Ww](hite|HITE).*"
    patron_tigre = r".*[Tt](iger|IGER).*"
    patron_gris = r".*[Gg](rey|REY).*"
    patron_lemon = r".*[Ll](emon|EMON).*"
    patron_bull = r".*[Bb](ull|ULL).*"

    try: 
        if re.findall(patron_blanco,col):
            return "White"
        elif re.findall(patron_tigre,col):
            return "Tiger"
        elif re.findall(patron_gris,col):
            return "Grey"
        elif re.findall(patron_lemon, col):
            return "Lemon"
        elif re.findall(patron_bull, col):
            return "Bull"
        else:
            return "Unspecified"
    except:
        return np.nan
    
```

```python
df["especie_cat"] = (df["species_"].apply(cambiar_especie))
```

### 2. Columna de age: es una columna de tipo string pero deber铆a ser de tipo integer. Adem谩s, en esta columna nos vamos a encontrar con algunos errores tipogr谩ficos, estos incluyen:

```python
df["age"].unique()
```

```python
def limpiar_edad (col):

    patron = "\d+"
    
    try:
        return int(col)

    except:
        if type(col) == str:
            if re.findall(patron,col):
                return int(re.findall(patron,col)[0])
        
        else:
            return np.nan
```

```python
df["age_limpio"] = (df["age"].apply(limpiar_edad))
```
